```python
#zipファイルからCSVファイルを取り出すコード
import os
import zipfile
import shutil

source_folder = R"C:\Users\umi08\Desktop\データ分析練習\【島根県】月毎の雨量データ"  # ZIPファイルが格納されているフォルダのパス
destination_folder = R"C:\Users\umi08\Desktop\データ分析練習\csv"  # CSVファイルを保存するフォルダのパス
```

```python
    # ソースフォルダ内のすべてのファイルを走査
    for filename in os.listdir(source_folder):
        # ZIPファイルのみ対象
        if filename.endswith(".zip"):
            zip_path = os.path.join(source_folder, filename)
            
            # ZIPファイルを解凍
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                # ZIPファイル内のファイルを走査
                for file in zip_ref.namelist():
                    # CSVファイルのみ対象
                    if file.endswith(".csv"):
                        # CSVファイルを出力先フォルダに抽出
                        zip_ref.extract(file, destination_folder)
                        # フォルダ構造を無視してCSVファイルを直接配置
                        extracted_path = os.path.join(destination_folder, os.path.basename(file))
                        
                        os.rename(os.path.join(destination_folder, file), extracted_path)
                        print(f"{file} を {destination_folder} に移動しました。")
    print("すべてのCSVファイルの抽出が完了しました。")
```
